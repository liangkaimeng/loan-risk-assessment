# 基本信息

## A_train数据集

```python
train_A_detect = detect(train_A).iloc[:, :4]
train_A_detect.head()
```

## B_train数据集

```python
train_B_detect = detect(train_B).iloc[:, :4]
train_B_detect.head()
```

## B_test数据集

```python
test_B_detect = detect(test_B).iloc[:, :4]
test_B_detect.head()
```

# 数据清洗

```python
train_A_enum_drop_columns = train_A_detect[train_A_detect["unique"] == 1].index.tolist()
train_B_enum_drop_columns = train_B_detect[train_B_detect["unique"] == 1].index.tolist()
test_B_enum_drop_columns = test_B_detect[test_B_detect["unique"] == 1].index.tolist()

enum_drop_slt_colnames = list()
enum_drop_colnames = train_A_enum_drop_columns + train_B_enum_drop_columns + test_B_enum_drop_columns
for colname in enum_drop_colnames:
    if colname not in enum_drop_slt_colnames:
        enum_drop_slt_colnames.append(colname)
        
train_A.drop([enum_drop_slt_colnames], axis=1, inplace=True)
train_B.drop([enum_drop_slt_colnames], axis=1, inplace=True)
test_B.drop([enum_drop_slt_colnames], axis=1, inplace=True)
```

```python
train_A_null_rate = train_A.isnull().sum() / train_A.shape[0]
train_A_null_drop_columns = train_A_null_rate[train_A_null_rate >= 0.9].index.tolist()

train_B_null_rate = train_B.isnull().sum() / train_B.shape[0]
train_B_null_drop_columns = train_B_null_rate[train_B_null_rate >= 0.9].index.tolist()

test_B_null_rate = test_B.isnull().sum() / train_B.shape[0]
test_B_null_drop_columns = test_B_null_rate[test_B_null_rate >= 0.9].index.tolist()

combine_colnames = train_A_null_drop_columns + train_B_null_drop_columns + test_B_null_drop_columns
null_drop_slt_colnames = list()
for colname in combine_colnames:
    if colname not in null_drop_slt_colnames:
        null_drop_slt_colnames.append(colname)
        
train_A.drop([null_drop_slt_colnames], axis=1, inplace=True)
train_B.drop([null_drop_slt_colnames], axis=1, inplace=True)
test_B.drop([null_drop_slt_colnames], axis=1, inplace=True)
```
